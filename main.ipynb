```python
import matplotlib.pyplot as plt
import numpy as np


def exponential_function(x):
    return 1 - 0.25 * np.exp(-2 * x)
    
def piecewise_function(x):
    return np.piecewise(x, [x <= 0, x > 0], [0, exponential_function])
def draw_dcf_graph():

    x_values = np.linspace(-4, 8, 1000) 
    y_values = piecewise_function(x_values)

    plt.plot(x_values, y_values, label='Piecewise Function')
    plt.title('Piecewise Function: Constant and Exponential')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.legend()
    plt.grid(True)
    plt.show()

draw_dcf_graph()
```

```python
def calculate_desired_answers(x):
    if x >= 0:
      return exponential_function(x)
    else:
      return 0
```

```python
#solving for P[X<=2]
calculate_desired_answers(2)
```

```python
#solving for P[X=0]
calculate_desired_answers(0)
```

```python
#solving for P[X<0]
# take a number less than 0 as an input such as -1
calculate_desired_answers(-1)
```

```python
#solving for P[0<=X<=6] = P[X<=6]-P[X<=2]
calculate_desired_answers(6) - calculate_desired_answers(2)
```

```python
#solving for P[X>10]= 1-P[X<=10]
1-calculate_desired_answers(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_exponential_distribution(scale, size=1000):
   
    uniform_samples = np.random.rand(size)
    exponential_samples = -scale * np.log(1 - uniform_samples)
    return exponential_samples

def plot_exponential_distribution_pdf(samples, scale):
    
    plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = (1 / scale) * np.exp(-x / scale)
    plt.plot(x, p, 'k', linewidth=2)

    title = "Fit results: scale = %.2f" % scale
    plt.title(title)
    plt.show()


scale_value = 2
samples = generate_exponential_distribution(scale_value)
plot_exponential_distribution_pdf(samples, scale_value)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_normal_distribution(mean, variance, size=1000):
   
    std_dev = np.sqrt(variance)
    normal_samples = mean + std_dev * np.random.randn(size)
    return normal_samples

def plot_normal_distribution_pdf(samples, mean, variance):
   
    plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-(x - mean)**2 / (2 * variance))
    plt.plot(x, p, 'k', linewidth=2)

    title = "Fit results: mean = %.2f,  variance = %.2f" % (mean, variance)
    plt.title(title)
    plt.show()
```

```python
mean_value = 5
variance_value = 4
samples = generate_normal_distribution(mean_value, variance_value)
plot_normal_distribution_pdf(samples, mean_value, variance_value)
```
